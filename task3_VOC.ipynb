```python


years = np.arange(2002, 2024)
bad_cols = ['Unnamed','VFlag','MDL','Sample Type','Sample ID#','Sample Date','Canister ID#','Sample Volume','NAPS ID','START TIME','DURATION']

voc = pd.DataFrame()
voc['Date'] = pd.date_range(start=f'1/1/2002', end='12/31/2023',freq='1D')
voc = voc.set_index('Date')

for yr in years:
  for iStn, stn in enumerate(stn_tags):
    if yr < 2014:
      sfile = f'S{stn}_VOC.XLS'
    elif yr >= 2014 and yr < 2016:
      sfile = f'S{stn}_VOC_{yr}.XLS'
    elif yr >= 2016 and yr < 2018:
      sfile = f'S{stn}_VOC_{yr}_EN.XLS'
    elif yr >= 2018:
      sfile = f'S{stn}_VOC_{yr}_EN.xlsx'

    filepath = os.path.join(naps_dir,str(yr),sfile)
    if os.path.isfile(filepath):
      print(f'Reading: {filepath}')

      #get skip row formatting
      if yr==2002 or ( yr >= 2008 and yr < 2014 ):
        skip=2
        dttag = 'Compounds'
      elif yr > 2002 and yr < 2008:
        skip=1
        dttag = 'Compounds'
      elif yr >= 2014 and yr < 2018:
        skip=0
        dttag = 'Sampling Date'
        sheet = 'Data'
      elif yr >= 2018:
        skip=8
        dttag = 'Sampling Date'
        sheet = 'VOC'

      xls = pd.ExcelFile(filepath)
      if len(xls.sheet_names) > 1:
        df = pd.read_excel(filepath, sheet_name=sheet, skiprows=skip,na_values=-999)
      else:
        df = pd.read_excel(filepath, skiprows=skip,na_values=-999)

      #remove unused columns
      for flag in bad_cols:
        df = df[[col for col in df.columns if flag not in col]]

      #set index to date
      df = df.rename(columns={dttag: 'Date'})
      df.index = pd.to_datetime(df.Date)
      df = df.drop(columns=['Date'])

      # Mark negative numerical values as NaN
      for col in df.select_dtypes(include=np.number):
          df.loc[df[col] <= 0, col] = np.nan

      numerical_cols = df.select_dtypes(include=np.number).columns
      df['tVOC'] = df[numerical_cols].sum(axis=1, skipna=True,min_count=1)

      #deal with duplicated indecies (multiple measurements on the same day)
      df_grouped = df.groupby(df.index)['tVOC'].mean()

      voc.loc[df_grouped.index, stn_ids[iStn]] = df_grouped.values


```
